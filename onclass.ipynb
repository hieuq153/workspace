```python
# Tính giai thừa của một số nguyên không âm (bao gồm cả số dương và == 0)
# n! = n * (n - 1)! với n > 0
# 0! = 1

def gt(n):
    if n <= 1:
        return 1
    return n * gt(n - 1)
    
n = int(input("Nhập một số nguyên không âm: "))
print(f"Giai thừa của {n} là: {gt(n)}")
```

```python
#Fibonacci là một dãy số bắt đầu bằng 0 và 1, sau đó mỗi số tiếp theo là tổng của hai số trước đó.

def fibonacci(n):
    if n == 1:
        return 0
    if n == 2:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
n = int(input("Nhập một số nguyên dương: "))
print(f"Số Fibonacci thứ {n} là: {fibonacci(n)}")
```

```python
#Merge sort là một thuật toán sắp xếp dựa trên ý tưởng của thuật toán sắp xếp nổi bọt (bubble sort). Tuy nhiên, thay vì so sánh và hoán đổi các phần tử liền kề, merge sort chia danh sách thành hai nửa, sắp xếp từng nửa riêng biệt, sau đó trộn hai nửa đã sắp xếp lại với nhau.


lst = [5, 2, 9, 1, 5, 6]
def merge(arr, st, mid, en):
    i = st
    j = mid
    tmp = []
    while i < mid and j <= mid:
        if arr[i] < arr[j]:
            tmp.append(arr[i])
            i += 1
        else:
            tmp.append(arr[j])
            j += 1
    while i < mid:
        tmp.append(arr[i])
        i += 1
    while j <= mid:
        tmp.append(arr[j])
        j += 1
    
    for i in range(len(tmp)):
        arr[i + st] = tmp[i]

def merge_sort(arr, st, en):
    if en - st <= 1:
        return
    mid = (st + en) // 2
    merge_sort(arr, st, mid)
    merge_sort(arr, mid + 1, en)
    merge(arr, st, mid, en)

merge_sort(lst, 0, len(lst)-1)
print(lst)
```
