W4A20

```python
n = int(input("Nhập n: "))
if n <= 0:
    print("Không phải")
else:
    a = n
    while a % 2 == 0:
        a //= 2
    if a == 1:
        print(n, "là số dạng 2^k")
    else:
        print(n, "không phải dạng 2^k")
```

W4A21 Tổng các chữ số

```python
n = abs(int(input()))
tong = 0
while n > 0:
    tong += n % 10
    n //= 10
print(tong)
```

W4A22

```python
n = abs(int(input()))
chan = 0
le = 0
if n == 0: chan = 1
while n > 0:
    if (n % 10) % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Chẵn:", chan, "Lẻ:", le)
```

```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print("k lớn nhất là:", k)
```

W4A25

```python
x = int(input())
if x != -1:
    max_v = x
    min_v = x
    while x != -1:
        if x > max_v: max_v = x
        if x < min_v: min_v = x
        x = int(input())
    print("Max:", max_v, "Min:", min_v)
```

W4A26

```python
a = int(input("Nhập A: "))
f1, f2 = 1, 1
b = 1
while f1 <= a:
    b = f1
    f = f1 + f2
    f1 = f2
    f2 = f
print("Số Fibonacci lớn nhất không vượt quá A là:", b)
```

W4A27

```python
s = input("Nhập chuỗi: ")
so_tu = 0
case = False

for kt in s:
    if kt != " ":
        if not case:
            so_tu += 1
            case = True
    else:
        case = False
print("Số từ trong chuỗi:", so_tu)
```

W4A29

```python
s = input()
kq = ""
for chu in s:
    if 'a' <= chu <= 'z':
        kq += chu.upper()
    elif 'A' <= chu <= 'Z':
        kq += chu.lower()
    else:
        kq += chu
print(kq)
```

```python
s = input() + ","
tong = 0
a = ""
for chu in s:
    if '0' <= chu <= '9':
        a += chu
    elif chu == "," and a != "":
        tong += int(a)
        a = ""
print(tong)
```
