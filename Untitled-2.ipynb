W8A1: Bài tập hình trụ

```python
class Cylinder:
    def __init__(self, bankinh, chieucao):
        self.r = bankinh
        self.h = chieucao

    def dientichtoanphan(self):
        return 2 * 3.14 * self.r * self.h + 2 * self.r ** 2 * 3.14
    def thetich(self):
        return 3.14 * self.r ** 2 * self.h
    def __str__(self):
        return f"Ban kinh: {self.r}, Chieu cao: {self.h}"
    
b, c = map(float, input().split())
cylinder = Cylinder(b, c)
print(cylinder)
print(f"Dien tich toan phan: {cylinder.dientichtoanphan():.2f}")
print(f"The tich: {cylinder.thetich():.2f}")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return f'(x={self.x}, y={self.y})'
    def gocToaDo(self):
        return self.x == 0 and self.y == 0
    def trucHoanh(self):
        return self.y == 0
    def trucTung(self):
        return self.x == 0
    def khoangCach(self, other):
        return round(((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5, 2)
x, y = map(int, input().split())
a = Point(x, y)
b = Point(0, 0)
print(a, a.gocToaDo(), a.trucHoanh(), a.trucTung(), a.khoangCach(b))
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def multiply(self):
        return self.a * self.b
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Không thể chia cho 0"
    def power(self):
        return self.a ** self.b
    def mod(self):
        if self.b != 0:
            return self.a % self.b
        else:  
            return "Không thể chia cho 0"
    def set_numbers(self, a, b):
        self.a = a
        self.b = b
    

a, b = map(float, input("Nhập hai số: ").split())
calc = Calculator(a, b)
choice = input()
while True:
    if choice == "1":
        print(calc.add())
    elif choice == "2":
        print(calc.subtract())
    elif choice == "3":
        print(calc.multiply())
    elif choice == "4":
        print(calc.divide())
    elif choice == "5":
        print(calc.power())
    elif choice == "6":
        print(calc.mod())
    elif choice == "7":
        while True:
            a, b = map(int, input("Nhập hai số mới: ").split())
            calc.set_numbers(a, b)
            break
    else:
        print("Bạn có muốn thoát không? (y/n)")
        thoat = input()
        if thoat in ["YES", "Yes", "yes"]:
            break
        else:
            choice = input("Chọn lại: ")
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def __str__(self):
        return f"Rectangle(width = {self.w}, height = {self.h})"
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2 * (self.w + self.h)
    def scale (self, k):
        self.w *= k
        self.h *= k
w, h, k = map(float, input("Nhập chiều rộng và chiều cao: ").split())
a = Rectangle(w, h)
print(a)
print(f"Tỷ lệ mở rộng: {k}")
a.scale(k)
print(f"Diện tích sau khi mở rộng: {a.area()}")
print(f"Chu vi sau khi mở rộng: {a.perimeter()}")
```
