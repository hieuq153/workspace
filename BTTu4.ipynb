W4A10

```python
n = int(input())
dd = 0
max = 0

for i in range(1, n + 1):
    m = i
    L = 1
    while m != 1:
        if m % 2 == 0:
            m //= 2
        else:
            m = m * 3 + 1
        L += 1
    
    if L > dd:
        dd = L
        max = i
print(max, dd)
```
