Bài 3: Nhập vào số nguyên dương a, in ra bảng cửu chương của a.

```python
n = int(input())
for i in range(1,10):
    print(f"{n} x {i} = {n*i}")
```

Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím.

```python
h = int(input())
for i in range(1,h+1):
    a = " " * (h - i)
    if i == 1:
        print(a + "*")
    elif i == h:
        print("*"* (2 * h - 1))
    else:
        b = " " * (2 * i - 3)
        print(a + "*" + b + "*")
```

Bài 5: Tính tổng 1 + 2 + 3 + ... + n

```python
n = int(input())
tong = 0
for i in range(1,n+1):
    tong += i
print(tong)
```

Bài 6: Ước dương của số n.

```python
n = int(input())
for i in range(1,n+1):
    if n % i == 0:
        print(i, end = " ")
```

Bài 7: Đếm số ước dương của số n.

```python
n = int(input())
s = 0
for i in range(1,n+1):
    if n % i == 0:
        s += 1
print(s)
```

Bài 8: Tìm ước dương chung của a và b.

```python
a, b = map(int, input().split())
for i in range(1, a+1):
    if a % i == 0 and b % i == 0:
        print(i, end = " ")
```

Bài 9: Xác định số chính phương.

```python
n = int(input())
a = False
for i in range(1,n+1):
    if i * i == n:
       a = True
       break
if a:
    print("là số chính phương")
else:
    print("không là số chính phương")
```

Bài 10: Nhập vào số nguyên dương, kiểm tra số hoàn hảo.

```python
a = int(input())
n = 0
for i in range(1,a):
    if a % i == 0:
        n += i
if n == a:
    print("là số hoàn hảo")
else:
    print("không là số hoàn hảo")
```
