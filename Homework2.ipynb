Bài 3: Tính tam giác , kiểm tra tam giác đều cân vuông

```python
a, b, c = map(float, input().split())
if a + b > c and b + c > a and a + c > b:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
    p = (a + b + c) / 2
    s = round((p * (p - a) * (p - b) * (p - c)) ** 0.5, 2)
    c = round(a + b + c , 2)
    print(f"Dien tich tam giac la {s}")
    print(f"Chu vi tam giac la {c}")
else:
    print("Ba canh khong tao thanh tam giac")
```

Bài 4: Chương trình giải ax + b = 0

```python
a, b = map(float, input("Nhập a, b cho ax + b = 0 ").split())
if a == 0:
    if b == 0:
        print("Phuong trinh co vo so nghiem")
    else:
        print("Phuong trinh vo nghiem")
else:
    x = round(-b/a, 2)
    print(f"Phuong trinh co nghiem x = {x}")
```

Bài 5: Giải phương trình bậc 2 ax^2 + bx + c = 0

```python
a, b ,c = map(int, input("Nhập a, b, c cho pt ax^2 + bx + c = 0 ").split())
print(f"Phuong trinh co dang {a}x^2 + {b}x + {c} = 0")
if a == 0:
    if b == 0:
        if c == 0:
            print("Phuong trinh co vo so nghiem")
        else:
            print("Phuong trinh vo nghiem")
    else:
        x = round(-c/b, 2)
        print(f"Phuong trinh co nghiem x = {x}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phuong trinh vo nghiem")
    elif delta == 0:
        x = round(-b/(2*a), 2)
        print(f"Phuong trinh co nghiem kep x = {x}")
    else:
        x1 = round((-b + delta**0.5) / (2*a), 2)
        x2 = round((-b - delta**0.5) / (2*a), 2)
        print(f"Phuong trinh co hai nghiem phan biet x1 = {x1} va x2 = {x2}")
```

Bài 8: Giải hệ phương trình bậc nhất 2 ẩn

```python
# ax + by = c
# dx + ey = f
a, b, c, d, e, f = map(float, input("Nhập a, b, c, d, e, f").split())
det = a*e - b*d
if det == 0:
    if a*f == c*d and b*f == c*e:
        print("He phuong trinh co vo so nghiem")
    else:
        print("He phuong trinh vo nghiem")
else:
    x = round((c*e - b*f) / det, 2)
    y = round((a*f - c*d) / det, 2)
    print(f"He phuong trinh co nghiem x = {x} va y = {y}")
```

Bài 10: Kiểm tra điểm thuộc đường tròn.

```python
#(x-a)^2 + (y-b)^2 = r^2

a, b, r = map(float, input("Nhập tọa độ tâm và bán kính đường tròn (a b r): ").split())
c, d = map(float, input("Nhập tọa độ điểm (c d): ").split())
if (c - a)**2 + (d - b)**2 == r**2:
    print("Điểm thuộc đường tròn")
else:
    print("Điểm không thuộc đường tròn")
```
