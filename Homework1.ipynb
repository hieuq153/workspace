Bài 2: Tính diện tích phần còn lại

```python
a = float(input("Nhập chiều rộng:"))
b = float(input("Nhập chiều dài:"))

r = a/2
S_hcn = a * b
S_tron = 3.14 * (r ** 2)
S = S_hcn - S_tron
print(S)
```

Bài 4: Kiểm tra kí tự alphabet:

```python
a = input("Nhập kí tự:")

if a.isalpha():
    print("Đây là chữ cái thuộc alphabet.")
else:
    print("Đây không phải là chữ cái thuộc alphabet.")
```

Bài 5: Tìm chữ cái liền trước.

```python
a = input("Nhập kí tự:")

if 65<ord(a)<=90:
    print(chr(ord(a)+31))
elif ord(a) == 65:
    print("z")
```

Bài 6: Tính diện tích tam giác

```python
a = float(input("Nhập cạnh 1:"))
b = float(input("Nhập cạnh 2:"))   
c = float(input("Nhập cạnh 3:"))

if a + b > c and a + c > b and b + c > a:
    print("Đây là 3 cạnh của một tam giác.")
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print("Diện tích tam giác là:", S)
else:
    print("Đây không phải là 3 cạnh của một tam giác.")
```

Bài 7: Tìm giá trị của vị trí thứ 5 và 9 của dãy kí tự.

```python
a = input("Dãy kí tự dài hơn hoặc bằng 20 kí tự:")

if len(a) >= 20:
    print(f"Kí tự thứ 5: {a[4]}")
    print(f"Kí tự thứ 9: {a[8]}")
else:
    print("Dãy kí tự không đủ 20 kí tự.")
```

Bài 8: Tính tiền điện:

```python
ho_ten = input("Nhập họ tên:")
a = float(input("Chỉ số tháng trước:"))
b = float(input("Chỉ số tháng này:"))

c = b - a

if c <= 50:
    d = c * 1984
elif c <= 100:
    d = 50 * 1984 + (c - 50) * 2050
elif c <= 200:
    d = 50 * 1984 + 50 * 2050 + (c - 100) * 2380
elif c <= 300:
    d = 50 * 1984 + 50 * 2050 + 100 * 2380 + (c - 200) * 2998
elif c <= 400:
    d = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (c - 300) * 3350
else:
    d = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (c - 400) * 3460

e = d * 1.08

print(f"Họ tên: {ho_ten}")
print(f"Tiền điện phải trả: {round(e)} đồng")
```
