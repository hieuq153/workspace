Bai 1: Tinh cuoc phi taxi

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        opening_fare = 11000
        fee_hai = 13000
        fee_hai_m = 10500
    elif car_type == 7:
        opening_fare = 12000
        fee_hai = 14000
        fee_hai_m = 12000

    fare_dist = 0
    if km <= 0.8:
        fare_dist = opening_fare
    elif km <= 20:
        fare_dist = opening_fare + (km - 0.8) * fee_hai
    else:
        fare_dist = opening_fare + (19.2 * fee_hai) + (km - 20) * fee_hai_m
    
    wait_fare = 0
    if wait_minutes > 5:
        blocks = ((wait_minutes - 5) // 5)
        wait_fare = blocks * 2000

    total_fare = fare_dist + wait_fare

    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        total_fare *= 1.2

    if km > 30 and not is_night:
        total_fare *= 0.9

    if total_fare < 50000 and km > 5:
        total_fare = 50000

    return int(total_fare)

car_type, start_hour, km, wait_minutes = map(int, input("Loại xe, Giờ bắt đầu, Số km, Thời gian chờ (phút): ").split())
print(f"Loại xe bạn chọn: {car_type}, Giờ bắt đầu: {start_hour}, Số km: {km}, Thời gian chờ: {wait_minutes} phút")
print(f"Tổng tiền cước: {smart_taxi_fare(car_type, start_hour, km, wait_minutes)}")
```
